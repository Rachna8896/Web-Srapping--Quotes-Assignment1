```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
Content=[]
for i in (1,2):
    url = f"http://quotes.toscrape.com/page/{i}/"
    response = requests.get(url)
    response = response.content 
    soup = BeautifulSoup(response,'html.parser')
    div = soup.find('div')
    contents = div.find_all('div',{'class':'quote'})
    for content in contents:
        quote = content.find('span')
        print(quote.text.strip())
        author = content.find('small')
        print(author.text.strip())
        tag = content.find('a',class_='tag')
        print(tag.text.strip())
        
        Content.append([quote,author,tag])
```

```python
df = pd.DataFrame(Content,columns = ['Quotes','Author','Tags'])
df
```

```python
df.to_csv('Quotes_Scrapped.csv')
```
